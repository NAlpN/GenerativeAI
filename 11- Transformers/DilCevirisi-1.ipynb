```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-tatoeba-en-tr")
```

```python
sentences = [
    "The weather is very nice today.",
    "I am learning how to use transformers for translation.",
]
```

```python
translations = translator(sentences)
```

```python
for translation in translations:
    print(translation['translation_text'])
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-tatoeba-en-tr")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-tatoeba-en-tr")
```

```python
text = "Hello, this is a test text. I am learning to translate languages."
```

```python
inputs = tokenizer.encode(text, return_tensors="pt", padding=True)

outputs = model.generate(inputs, max_length=40)
```

```python
translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(translated_text)
```
