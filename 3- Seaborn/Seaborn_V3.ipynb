```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data = sns.load_dataset("flights") 
```

```python
plot = sns.PairGrid(data) 
plot.map(plt.plot) 
plt.show()
```

```python
sns.scatterplot(x='year', y='passengers', data=data) 
plt.show()
```

```python
sns.lineplot(x='year', y='passengers', data=data) 
plt.show()
```

```python
sns.barplot(x='year', y='passengers', data=data) 
plt.show()
```

```python
sns.boxplot(x='year', y='passengers', data=data) 
plt.show()
```

```python
sns.violinplot(x='year', y='passengers', data=data) 
plt.show()
```

```python
sns.stripplot(x='year', y='passengers', data=data) 
plt.show()
```

```python
sns.distplot(data['passengers']) 
plt.show()
```

```python
sns.pairplot(data=data, hue='passengers') 
plt.show()
```

```python
sns.regplot(x='year', y='passengers', data=data) 
plt.show()
```
