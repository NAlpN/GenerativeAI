```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
```

```python
sns.distplot(data['total_bill'], kde=False, bins=30)
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=data)
plt.show()
```
