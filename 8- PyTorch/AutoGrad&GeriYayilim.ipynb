```python
import torch
```

```python
X = torch.tensor([1, 2, 3, 4], dtype = torch.float32)
Y = torch.tensor([2, 4, 6, 8], dtype = torch.float32)

w = torch.tensor(0.0, dtype = torch.float32, requires_grad = True)
```

```python
def forward(x):
  return w * x
```

```python
def loss(y, y_pred):
  return ((y_pred-y)**2).mean()
```

```python
print(f'EĞitim Öncesi Değer: f(5) = {forward(5).item():.3f}')
```

```python
for epoch in range(100):
  y_pred = forward(X)
  l = loss(Y, y_pred)
  l.backward()
  with torch.no_grad():
    w -= 0.01 * w.grad
  w.grad.zero_()

  if epoch % 10 == 0:
    print(f'epoch: {epoch+1}, loss: {l.item():.8f}, w: {w.item():.3f}')
```

```python
print(f'Eğitim Sonrası Değer: f(5) = {forward(5).item():.3f}')
```
