```python
tensor = torch.tensor([1, 2, 3])

print(tensor + 10)
print(tensor - 10)
print(tensor * 10)
print(tensor / 10)
```

```python
tensor = torch.tensor([1, 2, 3])

result_add = torch.add(tensor, 10)
print(result_add)

result_sub = torch.sub(tensor, 10)
print(result_sub)

result_mul = torch.mul(tensor, 10)
print(result_mul)

result_div = torch.div(tensor, 10)
print(result_div)
```

```python
tensor = torch.tensor([1, 2, 3])
torch.matmul(tensor, tensor)
```

```python
tensor = torch.tensor([[1, 2, 3], [1, 2, 3]])
print(tensor.shape)
print(tensor.T)
print(tensor.T.shape)
```

```python
tensor = torch.tensor([[1, 2, 3], [1, 2, 3]])

print(torch.min(tensor))
print(torch.max(tensor))
print(torch.mean(tensor.float()))
print(torch.sum(tensor))
```
