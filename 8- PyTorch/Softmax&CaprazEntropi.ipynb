```python
import torch
import torch.nn as nn
import numpy as np
```

```python
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x), axis = 0)
```

```python
x = np.array([2.0, 1.0, 0.1])
cikis = softmax(x)
print('Softmax Numpy: ', cikis)
```

```python
x = torch.tensor([2.0, 1.0, 0.1])
cikis = torch.softmax(x, dim = 0)
print('Softmax Torch: ', cikis)
```

```python
def capraz_entropi(actual, predicted):
  loss = -np.sum(actual * np.log(predicted))
  return loss
```

```python
Y = np.array([1, 0, 0])
```

```python
Y_pred_1 = np.array([0.7, 0.2, 0.1])
Y_pred_2 = np.array([0.1, 0.3, 0.6])
l1 = capraz_entropi(Y, Y_pred_1)
l2 = capraz_entropi(Y, Y_pred_2)

print(f'Loss1 Numpy: {l1:.4f}')
print(f'Loss2 Numpy: {l2:.4f}')
```

```python
loss = nn.CrossEntropyLoss()
```

```python
y = torch.tensor([0])
y_pred_1 = torch.tensor([[2.0, 1.0, 0.1]])
y_pred_2 = torch.tensor([[0.5, 2.0, 0.3]])

l1 = loss(y_pred_1, y)
l2 = loss(y_pred_2, y)

print(l1.item())
print(l2.item())
```

```python
_, predictions1 = torch.max(y_pred_1, 1)
_, predictions2 = torch.max(y_pred_2, 1)

print(predictions1)
print(predictions2)
```

```python
class NN2(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(NN2, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        out = self.linear1(x)
        out = self.relu(out)
        out = self.linear2(out)

        return out
```

```python
model = NN2(input_size=28*28, hidden_size=5, num_classes=3)
criterion = nn.CrossEntropyLoss()
```
