```python
import torch
```

```python
x = torch.randn(3, requires_grad = True)
print(x)
```

```python
y = x + 2
print(y)
```

```python
z = y * y * 2
z = z.mean()
print(z)
```

```python
v = torch.tensor([0.1, 1.0, 0.001], dtype = torch.float32)
z.backward()
print(x.grad)
```

```python
weights = torch.ones(4, requires_grad = True)

for epoch in range(3):
  model_cikis = (weights * 3).sum()
  model_cikis.backward()

  print(weights.grad)

  weights.grad.zero_()
```

```python

optimizer = torch.optim.SGD(weights, lr = 0.01)

optimizer.step()
optimizer.zero_grad()
```
