```python
import torch
import torch.nn as nn
```

```python
X = torch.tensor([[1], [2], [3], [4]], dtype=torch.float32)
Y = torch.tensor([[2], [4], [6], [8]], dtype=torch.float32)

n_s, n_f = X.shape
print(f'Örnek Sayısı: {n_s}, Özellik Sayısı: {n_f}')
X_test = torch.tensor([5], dtype=torch.float32)
```

```python
g_boyutu = n_f
c_boyutu = n_f
```

```python
model = nn.Linear(g_boyutu, c_boyutu)
```

```python
loss = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(1500):
    y_pred = model(X)

    l = loss(Y, y_pred)

    l.backward()

    optimizer.step()

    optimizer.zero_grad()

    if epoch % 150 == 0:
        [w, b] = model.parameters()
        print('epoch ', epoch+1, ': w = ', w[0][0].item(), ' loss = ', l)
```

```python
print(f'Eğitim Sonrası Değer: f(5) = {model(X_test).item():.3f}')
```
