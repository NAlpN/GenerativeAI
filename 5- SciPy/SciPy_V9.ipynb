```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
image = plt.imread('doga.jpeg')
```

```python
yeniden_image = ndimage.zoom(image, (0.5, 0.5, 1))
```

```python
dondur_image = ndimage.rotate(image, 45)
```

```python
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Görüntü")
```

```python
plt.subplot(1, 3, 2)
plt.imshow(yeniden_image)
plt.title("Yeniden Boyutlandırılmış Görüntü")
```

```python
plt.subplot(1, 3, 3)
plt.imshow(dondur_image)
plt.title("Döndürülmüş Görüntü")
```
