*OPTİMİZASYON*

```python
from scipy.optimize import minimize
```

```python
def rosen(x):
    return sum(100.0*(x[1:]-x[:-1]**2.0)**2.0 + (1-x[:-1])**2.0)
```

```python
x0 = [1.3, 0.7, 0.8, 2, 1.2]
```

```python
min = minimize(rosen, x0, method='nelder-mead', options={'xatol': 1e-8, 'disp': True})
```

```python
print(min.x)
```

*STATS*

```python
from scipy import stats
import numpy as np
```

```python
data = stats.norm.rvs(size=1000)
```

```python
mean = np.mean(data)
std_ = np.std(data)
```

```python
print("Ortalama:", mean)
print("Standart Sapma:", std_)
```

*FFT*

```python
import numpy as np
from scipy.fft import fft
```

```python
t = np.linspace(0, 1, 1000, endpoint=False)
frequencies = [10, 50, 100]
signal = np.sum([np.sin(2 * np.pi * freq * t) for freq in frequencies], axis=0)
```

```python
signal_fft = fft(signal)

print("Sinyalin Fourier Dönüşümü:")
print(signal_fft)
```
