```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def denklem(t, y):
    return t * np.sin(t)
```

```python
y0 = [0]
```

```python
t_span = (0, 10)
```

```python
sol = solve_ivp(denklem, t_span, y0)
```

```python
plt.plot(sol.t, sol.y[0])
plt.xlabel('Zaman')
plt.ylabel('Y')
plt.title('Diferansiyel Denklem Çözümü')
plt.grid(True)
plt.show()
```
