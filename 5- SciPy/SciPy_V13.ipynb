```python
from scipy.optimize import minimize
```

```python
def fonk(x):
    return (x[0] - 1) ** 2 + (x[1] - 2.5) ** 2
```

```python
initial_guess = [0, 0]
result = minimize(fonk, initial_guess)
```

```python
print("Optimal nokta:", result.x)
```

```python
from scipy.integrate import quad
from scipy.misc import derivative
```

```python
result, _ = quad(lambda x: x**2, 0, 1)
```

```python
result_derivative = derivative(lambda x: x**2, 1.0)
```

```python
print("Belirli Integral:", result)
print("Türev:", result_derivative)
```

```python
from scipy import stats
```

```python
data = [1, 2, 3, 4, 5]
```

```python
mean = np.mean(data)
std_dev = np.std(data)
```

```python
correlation_coefficient, p_value = stats.pearsonr(data, data)
```

```python
print("Ortalama:", mean)
print("Standart Sapma:", std_dev)
print("Korelasyon Katsayısı:", correlation_coefficient)
```

```python
import matplotlib.pyplot as plt
from scipy import ndimage, misc
```

```python
image = misc.ascent()
```

```python
filtered_image = ndimage.median_filter(image, size=3)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Orijinal Görüntü')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title('Filtrelenmiş Görüntü')

plt.show()
```

```python
from scipy.cluster.vq import kmeans, vq
import numpy as np
```

```python
data = np.random.rand(100, 2)
```

```python
centroids, distortion = kmeans(data, 3)
```

```python
clusters, _ = vq(data, centroids)
```

```python
print("Küme Merkezleri:", centroids)
print("Kümeleme Sonuçları:", clusters)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 4, 6])
```

```python
linear_regression = interp1d(x, y, kind='linear')
```

```python
plt.scatter(x, y, color='blue', label='Veri Noktaları')
plt.plot(x, linear_regression(x), color='red', label='Lineer Regresyon')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lineer Regresyon Analizi')
plt.legend()
```
