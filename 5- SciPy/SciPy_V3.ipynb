```python
from scipy.stats import linregress
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8,10])
y = np.array([1,3,5,7,9])
```

```python
egim, kesim, r_value, p_value, std_err = linregress(x,y)
```

```python
plt.scatter(x, y, label='Veri')
plt.plot(x, egim*x + kesim, color='green', label='Doğrusal Regresyon')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Doğrusal Regresyon')
plt.legend()
plt.show()
```

```python
print("Eğim:", egim)
print("Kesim:", kesim)
print("Korelasyon Katsayısı (r):", r_value)
print("P değeri:", p_value)
print("Standart Hata:", std_err)
```
