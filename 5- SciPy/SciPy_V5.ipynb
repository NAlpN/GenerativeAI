```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 9])
```

```python
def polinom(x, a, b, c):
    return a * x**2 + b * x + c
```

```python
params, _ = curve_fit(polinom, x, y)
```

```python
a, b, c = params
```

```python
x_reg = np.linspace(min(x), max(x), 100)
y_reg = polinom(x_reg, a, b, c)
```

```python
plt.scatter(x, y, label='Veri')
plt.plot(x_reg, y_reg, color='blue', label='Polinom Regresyonu')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
