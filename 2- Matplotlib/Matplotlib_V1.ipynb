```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 5])
y = np.array([0, 120])
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x_2, y_2, 'o')
plt.show()
```
