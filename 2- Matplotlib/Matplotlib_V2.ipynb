```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dunya.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.bar(df['Country'], df['Revenue'], color='green')
plt.xlabel('Ülke')
plt.ylabel('Gelir')
plt.title('Ülkelere Göre Gelir Dağılımı')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(df['Nüfus'], labels=df['Country'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Ülkelere Göre Nüfus Dağılımı')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Per person'], df['Nüfus'], color='red')
plt.xlabel('Kişi Başına Düşen Gelir')
plt.ylabel('Nüfus')
plt.title('Kişi Başına Düşen Gelir ve Nüfus Arasındaki İlişki')
plt.show()
```
